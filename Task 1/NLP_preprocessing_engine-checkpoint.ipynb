```python
# task 2

import re
def preprocess_text(text):
    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Remove URLs and emails
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (soooo → soo)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove special characters except spaces
    text = re.sub(r'[^a-zA-Z\s]', '', text)

    # Convert to lowercase
    text = text.lower()

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove very short tokens (<=2), keep exceptions
    exceptions = {"no", "not"}
    tokens = [word for word in tokens if len(word) > 2 or word in exceptions]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
#task 3

test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, cleaned = preprocess_text(sentence)
    results.append({
        "Original": sentence,
        "Tokens": tokens,
        "Cleaned": cleaned
    })

results
```

```python
#task 4

def token_analytics(tokens):
    if len(tokens) == 0:
        return 0, 0, 0

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(word) for word in tokens) / total

    return total, unique, round(avg_len, 2)


analytics_results = []

for item in results:
    tokens = item["Tokens"]
    total, unique, avg_len = token_analytics(tokens)

    analytics_results.append({
        "Sentence": item["Original"],
        "Total Tokens": total,
        "Unique Tokens": unique,
        "Avg Token Length": avg_len
    })

analytics_results
```

```python
#task 5

from collections import Counter

# Combine all tokens
all_tokens = []
for item in results:
    all_tokens.extend(item["Tokens"])

counter = Counter(all_tokens)

top_10 = counter.most_common(10)
least_5 = counter.most_common()[-5:]

print("Top 10 Frequent Words:", top_10)
print("Least 5 Frequent Words:", least_5)
```

```python
#task 6

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


pipeline_output = full_pipeline(test_sentences)
pipeline_output
```

```python
#task 7
def safe_preprocess(text):
    response = {
        "input": text,
        "tokens": [],
        "cleaned": "",
        "status": "",
        "message": ""
    }

    try:
        # Case 1: None input
        if text is None:
            response["status"] = "error"
            response["message"] = "Input is None"
            return response

        # Case 2: Non-string input
        if not isinstance(text, str):
            response["status"] = "error"
            response["message"] = "Input must be a string"
            return response

        # Case 3: Empty string
        if text.strip() == "":
            response["status"] = "warning"
            response["message"] = "Empty string provided"
            return response

        # Case 4: Only emojis or special chars
        if not any(char.isalnum() for char in text):
            response["status"] = "warning"
            response["message"] = "No valid textual content"
            return response

        # Case 5: Only numbers
        if text.isdigit():
            response["status"] = "warning"
            response["message"] = "Only numeric input"
            return response

        # Normal processing
        tokens, cleaned = preprocess_text(text)

        response["tokens"] = tokens
        response["cleaned"] = cleaned
        response["status"] = "success"
        response["message"] = "Processed successfully"

    except Exception as e:
        response["status"] = "error"
        response["message"] = str(e)

    return response
```

```python
test_inputs = [
    "",
    None,
    12345,
    "😂😂😂",
    "999999",
    "Valid text!!! 100%"
]

for inp in test_inputs:
    result = safe_preprocess(inp)
    print(result)
    print("-" * 50)
```
